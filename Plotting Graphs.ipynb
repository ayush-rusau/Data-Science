```python
# plot func x^3

import numpy as np
# x = np.array([1, 2, 3, 4, 5])
x = np.arange(0, 5, 0.1)
y = x ** 3
y2 = x**2
# plt.plot(x, y, color = "black", marker = "o", )
plt.plot(x, y, color = "black", linewidth = 5, label="x^3")
plt.plot(x, y2, color = "red", linewidth = 5, label = "x^2")
plt.legend()
# plt.axis([0, 10, 0, 150])
plt.text(2, 80, "Text", fontsize = 12)
plt.ylabel("y")
plt.xlabel("x")
plt.grid()
plt.title("matplotlob demo")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

employees=[61,71,79,91,93,89,90,94,99,128,118,114,124,131]
year=[2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018]
revenue=[39.79,44.28,51.12,60.42,58.44,62.48,69.94,73.72,77.85,86.83,93.58,85.32,89.95,110.36]
employees = employees
plt.scatter(year, revenue, s = employees)
plt.show()
for i in range(0, len(revenue)-1):
    if(revenue[i] > revenue[i-1] and revenue[i] > revenue[i+1]):
        print(year[i], revenue[i], employees[i])
```

```python
# Pie Graphs

labels = ["A", "B", "C", "D"]
sizes = [3, 4, 6, 2]
colors = ["blue","purple", "red", "pink"]
explode = [0.1, 0.2, 0, 0]
plt.title("Split among classes")
plt.pie(sizes, colors = colors, labels = labels, autopct="%.2f%%", explode = explode, counterclock = False, startangle = 100)
plt.axis("equal")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

company=np.array(['HP','Dell','Lenovo','Asus','Apple','Acer'])
sold=np.array([20000,43000,15000,17000,22000,13000])

plt.pie(sold, labels = company, autopct = "%3.2f%%")
plt.show()
sum = sold.sum()
for i in range(0, len(sold)):
    val = sold[i]/sum*100
    print(company[i], "%0.1f"%val)
```
