```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
d = {}
x = []
y = []

with open("amazon_jobs_dataset.csv") as file_obj:
    reader = csv.DictReader(file_obj, skipinitialspace = True)
    for row in reader:
        key = int(row['Posting_date'].split(' ')[-1])
        if(key in d):
            d[key] += 1
        else:
#             s.add(key)
            x.append(key)
            d[key] = 1
            
x.sort()
for i in x:
    y.append(d[i])
plt.plot(x, y, marker = "^")
plt.xlabel("Year")
plt.ylabel("Jobs")
plt.show()
for i in x:
    print(i, d[i])
```

```python
d = {}
month = ['January', 'February', 'March', 'April', 'May', 'June','July','August','September','October','November','December']
y = []
with open("amazon_jobs_dataset.csv") as file_obj:
    reader = csv.DictReader(file_obj, skipinitialspace = True)
    for row in reader:
#         print(row['Posting_date'].split(' ')[0])
        key = row['Posting_date'].split(' ')[0]
        if(key in d):
            d[key] += 1
        else:
            d[key] = 1
# print(d)       
for i in month:
    y.append(d[i])

plt.bar(month, y)
plt.xlabel("month")
plt.ylabel("Openings")
plt.xticks(rotation = 40)
plt.show()

for i in month:
    print(i, d[i])
```

```python
d = {}
x = []
with open("amazon_jobs_dataset.csv") as file_obj:
    reader = csv.DictReader(file_obj, skipinitialspace = True)
    for row in reader:
        arr=row['location'].split()
        if(len(arr) == 0):
            continue
        if(arr[0] == 'IN,'):
            key = arr[2]
            if(key in d):
                d[key] += 1
            else:
                d[key] = 1
                x.append(key)

sums = 0
for i in d:
    sums += d[i]
temp = {}
for i in d:
    temp[d[i]] = i
    
jobs = []
for i in temp:
    jobs.append(i)
jobs.sort()
label = []
for i in jobs:
    label.append(temp[i])
for i in range(0, len(jobs)):
    jobs[i] = (jobs[i]*100)/sums
    
plt.pie(jobs, labels = label, autopct = "%0.2f%%")
plt.show()
jobs.reverse()
label.reverse()
for i in range(0, len(jobs)):
    print(label[i], "%0.2f"%jobs[i])
```
