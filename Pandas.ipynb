```python
# count of freq of each type
import pandas as pd
column_name = ['a', 'b', 'c', 'd', 'e']
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", names = column_name)
a = iris.e
d = {}
# print(a)
for i in range(0, a.shape[0]):
    if(a[i] in d):
        d[a[i]] += 1
    else :
        d[a[i]] = 1
        
for i in d:
    print(d[i], end = " ")
```

```python
import pandas as pd
column_name = ['a', 'b', 'c', 'd', 'e']
iris = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", names = column_name)

i1 = iris[iris['e'] == 'Iris-setosa']
i2 = iris[iris['e'] == 'Iris-versicolor']
i3 = iris[iris['e'] == 'Iris-virginica']

k1 = i1.min(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-setosa")
k1 = i1.max(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-setosa")
k1 = i1.mean(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-setosa")

k1 = i2.min(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-versicolor")
k1 = i2.max(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-versicolor")
k1 = i2.mean(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-versicolor")

k1 = i3.min(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-virginica")
k1 = i3.max(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-virginica")
k1 = i3.mean(axis = 0)
print('%.2f'%k1['a'], '%.2f'%k1['b'], '%.2f'%k1['c'], '%.2f'%k1['d'], "Iris-virginica")
```
